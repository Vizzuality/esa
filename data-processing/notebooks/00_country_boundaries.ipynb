# Create country boundary layers

## Setup

### Library import

```python
from pathlib import Path

import geopandas as gpd
from data_processing.vector_processor import VectorProcessor
```

**Define variables**

```python
# Botswana
country = "Botswana"
BASE_PATH = Path("../data/raw/Water Management/Botswana/Vectors/")

input_file = BASE_PATH / "Botswana_Boundary.shp"
output_file = Path(str(BASE_PATH).replace("raw", "processed")) / "Botswana_Boundary.mbtiles"

layer_name = "Botswana_Boundary"
```

## Create country boundary `Shapfiles`

```python
countries = gpd.read_file("../data/raw/gadm_410-adm_0/gadm_410-adm_0.shp")
countries.head()
```

```python
gdf = countries[countries["COUNTRY"] == country]
gdf.to_file(input_file)
```

## Create vector layers from `Shapfiles`

```python
VectorProcessor(
    input_file=input_file,
    output_file=output_file,
    min_zoom=4,
    max_zoom=12,
    layer_name=layer_name,
    upload=True,
).process()
```
